```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import datetime
import random
import re
```

```python
random.seed(datetime.datetime.now())
```

```python
html = urlopen(f'https://www.nfl.com/')
bs = BeautifulSoup(html, 'html.parser')
for link in bs.find_all('a', href = re.compile('^[^https]')):
    if 'href' in link.attrs:
        print(link.attrs['href'])
```

```python
def getLinks(articleUrl):
    html = urlopen(f'https://www.nfl.com{articleUrl}')
    bs = BeautifulSoup(html, 'html.parser')
    return bs.find_all('a', href = re.compile('^[^https]'))
```

```python
links = getLinks('/')
counter = 0
while counter < 100 or len(links) == 0:
    newPage = links[random.randint(0, len(links) - 1)].attrs['href']
    print(newPage)
    counter += 1
```

avoiding the repetition

```python
pages = set()
def getLinks(pageURL):
    global pages
    try:
        html = urlopen(f'https://www.nfl.com{pageURL}')
    except Exception:
        print(f'https://www.nfl.com{pageURL} is an Invalid URL!')
        return('pageURL')
    
    bs = BeautifulSoup(html, 'html.parser')
    for link in bs.find_all('a', href = re.compile('^[^https]')):
        if 'href' in link.attrs:
            if link.attrs['href'] not in pages:
                newPage = link.attrs['href']
                print(newPage)
                pages.add(newPage)
                getLinks(newPage)
        

getLinks('')
    
```
